#What is Seaborn?

Seaborn is a Python data visualization library that is built on top of Matplotlib. It provides a high-level interface for creating visually appealing and informative statistical graphics. Seaborn simplifies the process of creating complex visualizations by providing a wide range of built-in functions and themes.<br>
Seaborn is designed to work seamlessly with Pandas data structures, making it an excellent choice for data analysis and exploration. It offers a variety of plot types, including scatter plots, line plots, bar plots, histograms, box plots, violin plots, heatmaps, and more. These plots can be customized and enhanced with various color palettes, themes, and styles.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Using the Seaborn library to create a relational plot (relplot) using the "flights" dataset <br>
The relplot function in Seaborn is used to create scatter plots or line plots to visualize the relationship between two variables.

```python
a = sns.load_dataset('flights')
sns.relplot(x = "passengers", y = "month", data = a)
```

The hue parameter allows us to add a third dimension to the plot by mapping a categorical variable to different colors.

```python
a = sns.load_dataset('flights')
sns.relplot(x = "passengers", y = "month",hue = "year", data = a)
```

By specifying kind = "line", Seaborn will generate a line plot where the x-axis represents the "time" variable and the y-axis represents the "tip" variable. Each data point will be connected by a line, creating a continuous line plot.

```python
b = sns.load_dataset("tips")
sns.relplot(x = "time", y="tip",data = b, kind = "line")
```

The catplot function in Seaborn is a higher-level function that can create various types of categorical plots, such as strip plots, swarm plots, box plots, and more.

```python
sns.catplot(x = "day", y = "total_bill",hue = "time", data = b)
```

kind = "violin" create a categorical plot using violin plots instead of the default strip plot.

```python
sns.catplot(x = "day", y = "total_bill", kind = "violin", data = b)
```

kind = "boxen" create a categorical plot using box plots instead of the default strip plot.

```python
sns.catplot(x = "day", y = "total_bill", kind = "boxen", data = b)
```

The **FacetGrid** class in Seaborn allows you to create a grid of subplots based on the unique values of a categorical variable. It provides a convenient way to visualize relationships between variables across different subsets of your data.

```python
a = sns.load_dataset("iris")
b = sns.FacetGrid(a, col = "species")
b.map(plt.hist, "sepal_length")
```

```python
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

The **PairGrid** class in Seaborn allows you to create a grid of subplots to visualize pairwise relationships between multiple variables. It provides a flexible way to explore the relationships between different variables in your dataset.

```python
sns.set(style = "dark")
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

**sns.despine(offset=10, trim=True)**: This line of code removes the spines (borders) from the plot using the despine() function from Seaborn. The offset parameter sets the distance between the spines and the plot, and the trim parameter trims the spines beyond the data range.

```python
sns.set(style = "white", color_codes = True)
a = sns.load_dataset("tips")
sns.boxplot(x = "day", y = "total_bill", data = a)
sns.despine(offset = 10, trim = True)
```

**c = sns.color_palette()**: This line of code calls the color_palette() function from Seaborn without any arguments. By default, this function returns a list of colors that form a color palette. <br>
**sns.palplot(c)**: This line of code uses the palplot() function from Seaborn to display the colors in the palette c. The palplot() function creates a plot where each color in the palette is represented by a vertical bar.

```python
c = sns.color_palette()
sns.palplot(c)
```
